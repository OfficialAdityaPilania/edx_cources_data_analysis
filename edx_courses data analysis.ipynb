```python
data['n_enrolled'] = data['n_enrolled'].str.replace(',', '')
data['n_enrolled'] = pd.to_numeric(data.n_enrolled)
data.sort_values('n_enrolled')
```

```python
data['price'] = data['price'].str.extract(r"(\d+)")
```

```python
data['price'] = pd.to_numeric(data.price)
```

```python
data.dtypes
```

```python
top_popular_courses = data[['title','n_enrolled']].sort_values(by='n_enrolled', ascending=False).head(10)
print(top_popular_courses)
```

```python
subject_counts = data['subject'].value_counts()
print(subject_counts)
```

```python
institution_counts = data['institution'].value_counts().head(10)
print(institution_counts)
```

```python
avg_price_by_subject = data.groupby('subject')['price'].mean().sort_values(ascending=False).head(10)
print(avg_price_by_subject)
```

```python
level_counts = data['Level'].value_counts()
print(level_counts)
```

```python
top_courses = data[['title', 'n_enrolled']].sort_values(by='n_enrolled', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='n_enrolled', y='title', data=top_courses, palette='viridis')
plt.title('Top 10 Most Enrolled Courses')
plt.xlabel('Enrollments')
plt.ylabel('Course Title')
plt.tight_layout()
plt.show()
```

```python
institution_counts = data['institution'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=institution_counts.values, y=institution_counts.index, palette='coolwarm')
plt.title('Top 10 Institutions by Number of Courses')
plt.xlabel('Number of Courses')
plt.ylabel('Institution')
plt.tight_layout()
plt.show()
```
